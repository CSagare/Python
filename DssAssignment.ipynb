```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a pandas dataframe
# df = pd.read_csv('customer_delinquency_data.csv')
df = pd.read_csv('Delinquent_customer.csv')

```

```python
df.info
```

```python
df.describe()
```

```python
# Create a histogram of the 'LIMIT_BAL' column
plt.hist(df['PERSONAL_MONTHLY_INCOME'], bins=50)
plt.xlabel('Credit Limit')
plt.ylabel('Frequency')
plt.title('Distribution of Credit Limits')
plt.show()
```

```python
# Create a box plot of the age column
sns.boxplot(x='TARGET_LABEL', y='AGE', data=df)
plt.title('Distribution of Age by Delinquency Status')
plt.xlabel('Non-delinquent (0) / Delinquent (1) ')
plt.ylabel('Age')
plt.show()
```

```python
# Create a bar chart of delinquent and non-delinquent customers
sns.countplot(x='TARGET_LABEL', data=df)
plt.xlabel('Delinquency Status')
plt.ylabel('Number of Customers')
plt.title('Number of Delinquent and Non-delinquent Customers')
plt.show()
```

```python
# Create a bar plot of the sex column
# sns.barplot(x='GENDER', y='APPLICATION_SUBMISSION_TYPE', data=df)
sns.barplot(x='GENDER', y='TARGET_LABEL', data=df)
plt.title('Delinquency Rate by Gender')
plt.xlabel('Sex (M = Male, F = Female)')
plt.ylabel('Delinquency Rate')
plt.show()
```

```python
# Create a heatmap of the correlation matrix
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Create a bar plot of the marriage column
sns.barplot(x='MARITAL_STATUS', y='TARGET_LABEL', data=df)
plt.title('Delinquency Rate by Marital Status')
# plt.xlabel('Marital Status (1 = Married, 2 = Single, 3 = Others)')
plt.ylabel('Delinquency Rate')
plt.show()
```

```python
# Create a bar plot of the Email column
sns.barplot(x='EMAIL', y='TARGET_LABEL', data=df)
plt.title('Delinquency Rate by Email')
plt.xlabel('(0= Dont Have Mail, 1 = Have mail)')
plt.ylabel('Delinquency Rate')
plt.show()
# Create a bar plot of the number column
sns.barplot(x='PROFESSIONAL_PHONE', y='TARGET_LABEL', data=df)
plt.title('Delinquency Rate by Number of Professional Phone')
plt.xlabel('(N= No, Y = Yes)')
plt.ylabel('Delinquency Rate')
plt.show()
```
